# Question 2 
(20 points)​: Write a SQL query that ranks all countries and leagues based on the total amount of total goals scored per game in the whole dataset. Sort them by the largest to the smallest amount of goals. Note: Read this carefully. The output of this query should be of the form:
Country | League Name |Total Goals Scored

```python
match_league = pd.read_sql("""SELECT m.country_id,lg.name,m.season,m.stage,m.date,m.match_api_id,m.home_team_api_id,m.away_team_api_id,m.home_team_goal,m.away_team_goal
                              FROM match m
                              JOIN league lg
                              ON m.league_id = lg.id""",conn)
match_league.to_sql("match_league", conn, if_exists="replace")
match_league.head(5000)

    
query = pd.read_sql("""SELECT
                       name,
                       SUM(home_team_goal+away_team_goal) 
                        FROM
                       match_league
                        GROUP BY name
                    ORDER BY SUM(home_team_goal+away_team_goal) DESC;""",conn)
query.head(5000)
    

```

# Question 3 
(20 points)​: Write a SQL query that ranks all teams by the average of all their attributes (not the players' attributes), sort them from best to worst. The output of this query should be of the form:
Team Long Name | Average of Attributes

```python


query = pd.read_sql("""SELECT team_long_name as TeamLongName, ROUND(AVG(((buildUpPlaySpeed + buildUpPlayDribbling + buildUpPlayPassing +chanceCreationPassing +chanceCreationCrossing +chanceCreationShooting +defenceAggression +defencePressure +defenceTeamWidth )/ 9)), 2) as TeamStats
        FROM Team
         JOIN Team_Attributes Y on Team.team_api_id= Y.team_api_id
        GROUP BY TeamLongName
        HAVING TeamStats NOT NULL
        ORDER BY TeamStats DESC;""",conn)
query.head(500000)


```

# Question 5 
(40 points)​: Write a SINGLE SQL query that finds the date that had the most goals scored on, per each different season and league. The output of this query should be of the form:
Date (dd/mm/yy) | Season | League Name | Goals scored


```python
query = pd.read_sql("""SELECT strftime('%d/%m/%Y', date) as 'Date(dd/mm/yy',season as Season, n.name LeagueName,
       MAX(ake.home_team_goal + ake.away_team_goal) as GoalsScored
        FROM Match ake
         JOIN League n on ake.league_id = n.id
            GROUP BY Season, LeagueName
            ORDER BY LeagueName, Season;""",conn)
query.head(500000)

```
